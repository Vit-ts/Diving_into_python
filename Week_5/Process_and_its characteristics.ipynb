## `Процесс и его характеристики` 

Характеристики процеса:
- Идентификация процесса, PID
- Обьем оперативной памяти
- Стек
- Список открытых файлов
- Ввод/вывод

```python
# простой Python процесс

import time,os

pid = os.getpid() # Идентификатор процесса, PID

while True:
    print(pid, time.time())
    time.sleep(2)
```

### ps aux


### sudo strace -p PID - системные вызовы процеса




### lsof -p PID - список запущеный файлов процесом


### python3 process.py > log.txt - вывод потока данных в файл

```python
with open('log.txt', 'r') as f:
    print(f.read())
```
